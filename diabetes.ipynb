```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_diabetes
```

```python
df = load_diabetes()
```

```python
df
```

```python
X = pd.DataFrame(df.data)
```

```python
X
```

```python
X.columns =  ['age',' sex','bmi','bp','s1','s2','s3','s4','s5','s6']
```

```python
Y = pd.DataFrame(df.target)
```

```python
Y.columns =["target"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size= 0.33, random_state= 42)
```

```python
X_train.corr()
```

```python
plt.figure(figsize=(12, 7))
sns.heatmap(X_test.corr(), annot=True)
```

```python
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor()
```

# hyper parameter tuning

```python
param = {
    "criterion" : ["squared_error", "friedman_mse", "absolute_error", "poisson"],
    "splitter" : ["best", "random"],
    "max_depth" : [1,2,3,4,5,6],
    "max_features" : ["sqrt", "log2", "auto"]
    
}
```

```python
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(estimator=model, param_grid=param, cv=5, n_jobs=-1,scoring="neg_mean_squared_error")
```

```python
grid.fit(X_train, Y_train)
```

```python
grid.best_params_
```

```python
y_pred = grid.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
print(r2_score(y_pred, Y_test))
print(mean_absolute_error(y_pred, Y_test))
print(mean_squared_error(y_pred, Y_test))
```

```python


model = DecisionTreeRegressor(criterion = 'poisson',  max_depth = 3, max_features= "log2", splitter="best")
model.fit(X_train, Y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(12, 7))
tree.plot_tree(model,filled=True)
```
