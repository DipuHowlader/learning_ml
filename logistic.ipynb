```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_classification
```

```python
X, Y =make_classification(n_samples=1000, n_features=10, n_classes=2, random_state= 42)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X.shape
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
logistic = LogisticRegression()
```

```python
logistic.fit(x_train, y_train)
y_pred = logistic.predict(x_test)
```

```python
y_pred
```

```python
logistic.predict_proba(x_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
score = accuracy_score(y_test, y_pred)
```

```python
cm = confusion_matrix(y_test, y_pred)
cr = classification_report(y_test, y_pred)
print(score)
print(cm)
print(cr)
```

## Hyperparameter tuning

```python
model = LogisticRegression()
```

```python
penalty = ['l1', 'l2', 'elasticnet']
c_values = [100, 10, 0.1, 1.0, 0.01]
solver = ['lbfgs', 'liblinear', 'newton-cg',' newton-cholesky', 'sag','saga']
param = dict(penalty= penalty, C = c_values, solver = solver)
```

```python
grid
```

```python
grid.fit(x_train, y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

## Logistic Regression for Imblance dataset

```python
from collections import Counter
from sklearn.datasets import make_classification
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x, y = make_classification(n_samples=10000, n_features=2, n_clusters_per_class=1, n_redundant=0, weights=[0.99], random_state=10)
```

```python
x
```

```python
Counter(y)
```

```python
import seaborn as sns
```

```python
sns.scatterplot(pd.DataFrame(x)[0], pd.DataFrame(x)[1], hue=y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
logisticregression = LogisticRegression()
```

```python
penalty = ['l1','l2', 'elasticnet']
c = [100, 10, 1.0, 0.1, 0.01]
```

```python
solver = ['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga']
```

```python
class_weight = [{0: w, 1: y} for w in [1, 10, 50, 100] for y in [1, 10, 50, 100]]
```

```python
class_weight
```

```python
c_values = [100, 10, 0.1, 1.0, 0.01]
```

```python
params = dict(penalty = penalty, C= c_values, solver = solver, class_weight = class_weight)
```

```python
params
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold
cv = StratifiedKFold()
```

```python
grid = GridSearchCV(estimator=logisticregression, param_grid= params, scoring='accuracy', cv = cv)
```

```python
grid.fit(x_train, y_train)
```

```python
grid.best_params_
```

```python
y_pred = grid.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
score = accuracy_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)
cr = classification_report(y_test, y_pred)
print(score)
print(cm)
print(cr)
```
